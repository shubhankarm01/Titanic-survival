```python
import re

# to handle datasets
import pandas as pd
import numpy as np

# for visualization
import matplotlib.pyplot as plt

# to divide train and test set
from sklearn.model_selection import train_test_split

# feature scaling
from sklearn.preprocessing import StandardScaler

# to build the models
from sklearn.linear_model import LogisticRegression

# to evaluate the models
from sklearn.metrics import accuracy_score, roc_auc_score

# to persist the model and the scaler
import joblib

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

## Prepare the data set

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/16826755/phpMYEkMl')

data.head()
```

```python
# replace interrogation marks by NaN values

data = data.replace('?', np.nan)
```

```python
# retain only the first cabin if more than
# 1 are available per passenger

def get_first_cabin(row):
    try:
        return row.split()[0]
    except:
        return np.nan
    
data['cabin'] = data['cabin'].apply(get_first_cabin)
```

```python
# extracts the title (Mr, Ms, etc) from the name variable

def get_title(passenger):
    line = passenger
    if re.search('Mrs', line):
        return 'Mrs'
    elif re.search('Mr', line):
        return 'Mr'
    elif re.search('Miss', line):
        return 'Miss'
    elif re.search('Master', line):
        return 'Master'
    else:
        return 'Other'
    
data['title'] = data['name'].apply(get_title)
```

```python
# cast numerical variables as floats

data['fare'] = data['fare'].astype('float')
data['age'] = data['age'].astype('float')
```

```python
# drop unnecessary variables

data.drop(labels=['name','ticket', 'boat', 'body','home.dest'], axis=1, inplace=True)

# display data
data.head()
```

```python
# save the data set

data.to_csv('titanic.csv', index=False)
```

## Separate data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop('survived', axis=1),  # predictors
    data['survived'],  # target
    test_size=0.2,  # percentage of obs in test set
    random_state=0)  # seed to ensure reproducibility

X_train.shape, X_test.shape
```

## Feature Engineering

### Extract only the letter (and drop the number) from the variable Cabin

```python
X_train['cabin'] = X_train['cabin'].apply(lambda x: x[0] if x != 'nan' else 'Missing')
X_test['cabin'] = X_test['cabin'].apply(lambda x: x[0] if x != 'nan' else 'Missing')

# data['cabin'] = data['cabin'].apply(lambda x: x[0] if x != 'nan' else 'Missing')
```

### Fill in Missing data in numerical variables:

- Add a binary missing indicator
- Fill NA in original variable with the median

```python
for column in vars_num:
    X_train[column + '_missing'] = np.where(X_train[column].isna(), 1, 0)
    X_train[column] = np.where(X_train[column].isna(), X_train[column].median(), X_train[column])
    
    X_test[column + '_missing'] = np.where(X_test[column].isna(), 1, 0)
    X_test[column] = np.where(X_test[column].isna(), X_test[column].median(), X_test[column])


# for column in vars_num:
#     data[column + '_missing'] = np.where(data[column].isna(), 1, 0)
#     data[column] = np.where(data[column].isna(), data[column].median(), data[column])
```

### Replace Missing data in categorical variables with the string **Missing**

```python
for column in vars_cat:
    X_train[column] = np.where(X_train[column] == 'nan', 'Missing', X_train[column])
    X_test[column] = np.where(X_test[column] == 'nan', 'Missing', X_test[column])
    
# for column in vars_cat:
#     data[column] = np.where(data[column] == 'nan', 'Missing', data[column])
```

### Remove rare labels in categorical variables

- remove labels present in less than 5 % of the passengers

```python
def rare_labels(column):
    label_freq = data.groupby(column)[column].count()/len(data)
    return(label_freq[label_freq < 0.05].index)
```

```python
for column in vars_cat:
    X_train[column] = np.where(X_train[column].isin(list(rare_labels(column))), 'Rare', X_train[column])
    X_test[column] = np.where(X_test[column].isin(list(rare_labels(column))), 'Rare', X_test[column])
    
#     X_train = X_train[~ X_train[column].isin(list(rare_labels(column)))]
```

```python
X_train.head()
```

### Perform one hot encoding of categorical variables into k-1 binary variables

- k-1, means that if the variable contains 9 different categories, we create 8 different binary variables
- Remember to drop the original categorical variable (the one with the strings) after the encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown = 'ignore')
temp_df = pd.DataFrame(ohe.fit_transform(X_train[vars_cat]).toarray(), columns = ohe.get_feature_names_out(), index = X_train.index) 
X_train = pd.concat([X_train.drop(vars_cat, axis = 1), temp_df], axis = 1)

temp_df = pd.DataFrame(ohe.transform(X_test[vars_cat]).toarray(), columns = ohe.get_feature_names_out(), index = X_test.index) 
X_test = pd.concat([X_test.drop(vars_cat, axis = 1), temp_df], axis = 1)
```

### Scale the variables

- Use the standard scaler from Scikit-learn

```python
X_train
```

```python
StandardScaler().fit_transform(X_train)[0]
```

```python
scaler = StandardScaler()
X_train[vars_num] = scaler.fit_transform(X_train[vars_num])
X_test[vars_num] = scaler.transform(X_test[vars_num])
```

## Train the Logistic Regression model

- Set the regularization parameter to 0.0005
- Set the seed to 0

```python
lr = LogisticRegression(random_state = 0, tol = 0.0005)
lr.fit(X_train, y_train)
```

## Make predictions and evaluate model performance

Determine:
- roc-auc
- accuracy

**Important, remember that to determine the accuracy, you need the outcome 0, 1, referring to survived or not. But to determine the roc-auc you need the probability of survival.**

```python
pred = lr.predict(X_test)
accuracy_score(y_test, pred)
```

```python
pred_prob_fr_posit = lr.predict_proba(X_test)[:,1]


from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, pred_prob_fr_posit)
```
